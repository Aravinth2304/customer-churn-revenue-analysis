```python
import pandas as pd
import numpy as np

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
df.info()
```

```python
df["subscription_plan"] = np.where(
    df["monthly_fee"] < 50, "Basic",
    np.where(df["monthly_fee"] < 80, "Standard", "Premium")
)
```

```python
df["subscription_plan"].value_counts()
```

```python
columns_to_keep = [
    "customer_id",
    "gender",
    "senior_citizen",
    "tenure_months",
    "monthly_fee",
    "total_charges",
    "payment_method",
    "churned",
    "subscription_plan"
]

df = df[columns_to_keep]
```

```python
df = df.rename(columns={
    "customerID": "customer_id",
    "SeniorCitizen": "senior_citizen",
    "tenure": "tenure_months",
    "MonthlyCharges": "monthly_fee",
    "TotalCharges": "total_charges",
    "PaymentMethod": "payment_method",
    "Churn": "churned"
})
```

```python
df
```

```python
df["gender"] = df["gender"].str.capitalize()
```

```python
df["payment_method"] = df["payment_method"].replace({
    "Electronic check": "Electronic",
    "Mailed check": "Mailed",
    "Bank transfer (automatic)": "Bank Transfer",
    "Credit card (automatic)": "Credit Card"
})
```

```python
df["total_charges"] = pd.to_numeric(df["total_charges"], errors="coerce")
```

```python
df["senior_citizen"] = df["senior_citizen"].astype(int)
df["tenure_months"] = df["tenure_months"].astype(int)
df["monthly_fee"] = df["monthly_fee"].astype(float)
df["total_charges"] = df["total_charges"].astype(float)
```

```python
df["total_charges"] = df["total_charges"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
df.duplicated(subset="customer_id").sum()
df["churned"].value_counts(normalize=True)
```

```python
df = df[
    [
        "customer_id",
        "gender",
        "senior_citizen",
        "tenure_months",
        "monthly_fee",
        "total_charges",
        "payment_method",
        "subscription_plan",
        "churned"
    ]
]

df.to_csv(
    "final_telco_churn_cleaned.csv",
    index=False,
    encoding="utf-8"
)
```
